```python
import cv2
import numpy as np

# Load the image
image_path = './cnn-1.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# List of filter matrices for different effects
filter_matrices = [
    np.array([[-1, -1, -1],
              [-1,  9, -1],
              [-1, -1, -1]]),  # Enhanced Edge Detection

    np.array([[0, -1, 0],
              [-1, 5, -1],
              [0, -1, 0]]),    # Sharpen

    np.array([[1, 2, 1],
              [2, 4, 2],
              [1, 2, 1]]),     # Blur

    np.array([[-2, -1, 0],
              [-1,  1, 1],
              [0,   1, 2]]),   # Emboss

    np.array([[0, 1, 0],
              [1, -4, 1],
              [0, 1, 0]]),     # Laplacian
]

# Apply each filter and display the results
for idx, filter_matrix in enumerate(filter_matrices):
    filtered_image = cv2.filter2D(image, -1, filter_matrix)
    cv2.imshow(f'Filtered Image {idx+1}', filtered_image)

# Display the original image
cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = './cnn-1.jpeg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# List of filter matrices along with their names
filters = [
    (np.array([[1, 2, 1],
               [2, 4, 2],
               [1, 2, 1]]), 'Blur'),

    (np.array([[0, -1, 0],
               [-1, 5, -1],
               [0, -1, 0]]), 'Sharpen'),

    (np.array([[-2, -1, 0],
               [-1,  1, 1],
               [0,   1, 2]]), 'Emboss'),

    (np.array([[0, 1, 0],
               [1, -4, 1],
               [0, 1, 0]]), 'Laplacian'),

    (np.array([[-1, -1, -1],
               [-1,  9, -1],
               [-1, -1, -1]]), 'Enhanced Edge Detection'),
]

# Display the original RGB image and apply and display each filter
plt.figure(figsize=(12, 8))

# Display the original RGB image
plt.subplot(2, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Apply and display each filter
for idx, (filter_matrix, filter_name) in enumerate(filters):
    filtered_image = cv2.filter2D(image_rgb, -1, filter_matrix)
    plt.subplot(2, 3, idx + 2)
    plt.imshow(filtered_image)
    plt.title(filter_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = 'cnn-1.jpeg'
image_rgb = cv2.imread(image_path)
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)

# List of filter matrices along with their names
filters = [
    (np.array([[-1, -1, 0],
               [-1,  0, 1],
               [0,   1, 1]]), 'Emboss'),

    (np.array([[1, 1, 1],
               [1, 1, 1],
               [1, 1, 1]]) / 9.0, 'Blur'),

    (np.array([[0, -1, 0],
               [-1, 5, -1],
               [0, -1, 0]]), 'Sharpen'),

    (np.array([[0, 1, 0],
               [1, -4, 1],
               [0, 1, 0]]), 'Laplacian'),

    (np.array([[-1, -1, -1],
               [-1,  9, -1],
               [-1, -1, -1]]), 'Enhanced Edge Detection'),
]

# Display the original grayscale image and apply/display each filter
plt.figure(figsize=(15, 10))

# Display the original grayscale image
plt.subplot(3, 3, 1)
plt.imshow(image_rgb)
plt.title('Original RGB Image')
plt.axis('off')

# Convert to grayscale
plt.subplot(3, 3, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Apply and display each filter for grayscale image
for idx, (filter_matrix, filter_name) in enumerate(filters):
    filtered_image = cv2.filter2D(image_gray, -1, filter_matrix)
    plt.subplot(3, 3, idx + 3)
    plt.imshow(filtered_image, cmap='gray')
    plt.title(filter_name + ' (Grayscale)')
    plt.axis('off')

# Apply and display each filter for RGB image
for idx, (filter_matrix, filter_name) in enumerate(filters):
    filtered_image_rgb = cv2.filter2D(image_rgb, -1, filter_matrix)
    plt.subplot(3, 3, idx + 6)
    plt.imshow(filtered_image_rgb)
    plt.title(filter_name + ' (RGB)')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image in grayscale
image_path = './cheems.webp'
gray_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Load an image in RGB
rgb_image = cv2.imread(image_path)

# Define different filters
filters = [
    ("Emboss", np.array([[-2, -1, 0],
                         [-1, 1, 1],
                         [0, 1, 2]])),

    ("Sharpened", np.array([[0, -1, 0],
                            [-1, 5, -1],
                            [0, -1, 0]])),
    ("Blurred", np.array([[1, 2, 1],
                           [2, 4, 2],
                           [1, 2, 1]]) / 16.0,),

    ("Dilated Edges", np.array([[-1, -1, -1],
                               [-1, 8, -1],
                               [-1, -1, -1]])),

    ("Edges", np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])),

]

# Apply filters and display results
plt.figure(figsize=(12, 18))
for i, (filter_name, filter_matrix) in enumerate(filters, 1):
    filtered_gray = cv2.filter2D(gray_image, -1, filter_matrix)
    filtered_rgb = cv2.filter2D(rgb_image, -1, filter_matrix)

    plt.subplot(5, 2, i * 2 - 1)
    plt.imshow(filtered_gray, cmap='gray')
    plt.title(f'{filter_name} (Grayscale)')
    plt.axis('off')

    plt.subplot(5, 2, i * 2)
    plt.imshow(cv2.cvtColor(filtered_rgb, cv2.COLOR_BGR2RGB))
    plt.title(f'{filter_name} (RGB)')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image_path = 'cnn-1.jpeg'
image = cv2.imread(image_path)

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Define different filters
identity_filter = np.array([[0, 0, 0],
                            [0, 1, 0],
                            [0, 0, 0]])

blur_filter = np.array([[1, 1, 1],
                        [1, 1, 1],
                        [1, 1, 1]]) / 9.0

sharpen_filter = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

edge_filter = np.array([[-1, -1, -1],
                        [-1, 8, -1],
                        [-1, -1, -1]])

emboss_filter = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]])

# Apply filters using cv2's filter2D function
filtered_images = {
    'Identity': cv2.filter2D(gray_image, -1, identity_filter),
    'Blur': cv2.filter2D(gray_image, -1, blur_filter),
    'Sharpen': cv2.filter2D(gray_image, -1, sharpen_filter),
    'Edge': cv2.filter2D(gray_image, -1, edge_filter),
    'Emboss': cv2.filter2D(gray_image, -1, emboss_filter),
}

# Display filtered grayscale images
plt.figure(figsize=(15, 10))
for i, (title, img) in enumerate(filtered_images.items()):
    plt.subplot(2, 3, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.show()

# Apply filters to RGB image channels separately
r_channel, g_channel, b_channel = cv2.split(image)

filtered_r = cv2.filter2D(r_channel, -1, edge_filter)
filtered_g = cv2.filter2D(g_channel, -1, sharpen_filter)
filtered_b = cv2.filter2D(b_channel, -1, emboss_filter)

filtered_rgb_image = cv2.merge((filtered_r, filtered_g, filtered_b))

# Display filtered RGB image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(filtered_rgb_image, cv2.COLOR_BGR2RGB))
plt.title('Filtered RGB Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = "cnn-1.jpeg"
image = cv2.imread(image_path, cv2.COLOR_BAYER_GRBG2GRAY)

filters = [
    ("Original", image),
    ("Blurred", cv2.GaussianBlur(image, (5, 5), 0)),
    ("Edges", cv2.Canny(image, 100, 200)),
    ("Emboss", cv2.filter2D(image, cv2.CV_64F, np.array([[0, -1, -1], [1, 0, -1], [1, 1, 0]]))),
    ("Sharpened", cv2.filter2D(image, cv2.CV_64F, np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]))),
    ("Dilated Edges", cv2.dilate(cv2.Canny(image, 100, 200), None, iterations=2)),

]

plt.figure(figsize=(12, 8))
for i, (filter_name, filtered_image) in enumerate(filters):
    plt.subplot(2, 3, i + 1)
    plt.imshow(filtered_image, cmap='gray')
    plt.title(filter_name)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
