```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)
```

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)
```

```python
def insert(trie, key, value):
    if key:
        first, rest = key[0], key[1:]
        if first not in trie:
            trie[first] = {}
        insert(trie[first], rest, value)
    else:
        trie['value'] = value
```

```python
# Space-Time Tradeoffs
# A simple text retrieval system

def raw(file):
    contents = open(file).read()
    contents = re.sub(r'<.*?>',  ' ', contents)
    contents = re.sub('\s+', ' ', contents)
    return contents
```

```python
def snippet(doc, term): #buggy
    text = ' '*30 + raw(doc) + ' '*30
    pos = text.index(term)
    return text[pos-30:pos+30]
```

```python
print("PREDICTING YOUR FUTURE BY AI...") 
future = nltk.corpus.movie_reviews.abspaths() 
idx = nltk.Index((w, f) for f in future for w in raw(f).split()) 

query = '' 
count = 0
while query != 'quit':
    query = input("Query >>> ") 
    if query in idx: 
        for doc in idx[query]: 
            if (snippet(doc, query)):
                count += 1
        print(f'{count} Possibilities found')
    else:
        print("Not found")
```

```python
colors = 'rgbcmyk'
def bar_chart(categories, words, counts):
    import pylab
    ind = pylab.arange(len(words))
    width = 1 / (len(categories) + 1)
    bar_groups = []
    for c in range(len(categories)):
        bars = pylab.bar(ind+c*width, counts[categories[c]], 
                         width, color=colors[c%len(colors)])
        bar_groups.append(bars)
    pylab.xticks(ind+width, words)
    pylab.legend([b[0] for b in bar_groups], categories, loc='upper left')
    pylab.ylabel('Frequency')
    pylab.title('Frequency of Six Modal Verbs by Genre')
    pylab.show()
```

```python
genres = ['news', 'religion', 'hobbies', 'government', 'adventure']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfdist = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in genres
    for word in nltk.corpus.brown.words(categories = genre)
    if word in modals
)
```

```python
counts = {}
for genre in genres:
    counts[genre] = [cfdist[genre][word] for word in modals]
bar_chart(genres, modals, counts)
```

```python
import networkx as nx 
import matplotlib 
from networkx.drawing.nx_agraph import graphviz_layout
from nltk.corpus import wordnet as wn
```

```python
def traverse(graph, start, node):
    graph.depth[node.name] = node.shortest_path_distance(start)
    for child in node.hyponyms():
        graph.add_edge(node.name, child.name)
        traverse(graph, start, child)
```

```python
def hyponym_graph(start):
    G = nx.Graph()
    G.depth = {}
    traverse(G, start, start)
    return G
```

```python
def graph_draw(graph):
    nx.draw(graph,node_size = [16 * graph.degree(n) for n in graph],
            node_color = [graph.depth[n] for n in graph], 
            with_labels = False)
    matplotlib.pyplot.show()
```

```python
dog = wn.synset('dog.n.01')
graph = hyponym_graph(dog)
graph_draw(graph)
```
